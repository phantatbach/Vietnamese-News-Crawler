```python
import re 
def get_category_from_url(url):
    items = url.split('/')
    category = None
    if len(items) >= 4:
        category = re.sub(r'-p[0-9]+', '', items[3])
    return category
get_category_from_url('https://vnexpress.net/the-gioi-p4')
```
